# Import libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sc
import pandas as pd
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
# line plot
sns.lineplot(x='sepal_length',y='sepal_width',data=df)
plt.title('sepal length V/S sepal width')
sns.set_style('dark')
sns.set_style(style=None,rc=None)
```

```python
# bar plot
sns.barplot(x='species',y='sepal_width',data=df)
plt.title('species V/S sepal width')
```

```python
sns.barplot(x='species',y='petal_length',data=df)
plt.title('species V/S petal_length')
```

```python
kashti=sns.load_dataset('titanic')
kashti
```

```python
sns.barplot(x='sex',y='alone',hue='who',data=kashti,order=['female','male'],color='red')
plt.title('sex V/S alive')
```

```python
# Horizental Bar plot
sns.barplot(x='fare',y='class',hue='who',data=kashti,color='blue')
plt.title('sex V/S alive')
```

```python
# boxplot
sns.set_style(style='whitegrid')
sns.boxplot(x='class',y='fare',data=kashti)
```

```python
sns.set_style(style='whitegrid')
sns.boxenplot(x='class',y='fare',data=kashti)
```

```python
sns.heatmap(kashti.corr(),annot=True)
```
