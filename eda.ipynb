```python
import seaborn as sns
sns.get_dataset_names()
```

```python
df=sns.load_dataset('diamonds')
df
```

```python
df.sample(10,random_state=999)
```

```python
# EDA
print('----describe---')
print(df.describe())
print('----info---')
print(df.info())
print('----columns---')
print(df.columns)
print('----number of unique values---')
print(df.nunique())
print('----count---')
print(df.count)
print('----size---')
print(df.size)
print('----shape---')
print(df.shape)
```

```python
# check null values
df.isnull().sum()
```

```python
# check datatype of our data
df.dtypes
```

```python
# change the column name of x,y and z
df.rename(columns = {'x':'length', 'y':'width', 'z': 'Depth'}, inplace = True)
df.head()
```

```python
# convert categorical data into numeric data
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['cut']=label.fit_transform(df['cut'])
df['clarity']=label.fit_transform(df['clarity'])
df['color']=label.fit_transform(df['color'])
df.head()
```

```python
# check co-relation between variables
df.corr()
```
